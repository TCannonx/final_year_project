```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
```

```python
results = pd.read_csv('Results.csv')
results = results.dropna(axis=0)
results.shape
```

```python
def check(results):    
    if (results['choice'] == results['improved_index']):
        return 1
    else:
        return 0
```

```python
results = results.assign(result=results.apply(check, axis=1))
```

```python
import matplotlib.pyplot as plt

# Data
values = results['result'].to_list()

# Count occurrences of each value
value_counts = {0: 0, 1: 0}
for value in values:
    value_counts[value] += 1

# Pie chart labels
labels = ['Original Model', 'Improved Model']

# Pie chart values
sizes = [value_counts[0], value_counts[1]]

# Set colors for the chart
colors = ['#0476D0', '#DA70D6']

# Set font properties
font = {'family': 'Arial',
        'weight': 'normal',
        'size': 12}

# Create the pie chart
fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%',
       startangle=90, wedgeprops={'edgecolor': 'white'})

# Set aspect ratio to be equal so that pie is drawn as a circle
ax.axis('equal')

# Add a title
plt.title("Value Distribution Between Participant's Choices Origin / Imporved Model", fontdict={'fontweight': 'bold', 'fontsize': 14})

# Set font properties for labels
plt.setp(ax.get_xticklabels(), fontfamily='Arial', fontsize=10)
plt.setp(ax.get_yticklabels(), fontfamily='Arial', fontsize=10)

# Adjust layout to fit labels properly
plt.tight_layout()

# Display the chart
plt.show()
```

```python
null_hypothesis_value = .5

# Perform single-sample t-test
t_statistic, p_value = stats.ttest_1samp(results['result'], null_hypothesis_value)

# Print the t-statistic and p-value
print("T-statistic:", t_statistic)
print("P-value:", p_value)
```
